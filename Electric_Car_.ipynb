**Relationship between Power with segment and seats**


1.  This chart shows how the distribution of electric cars varies across different segments for each powertrain type.
2.  This chart allows you to see how the number of seats varies among different powertrain types. we can determine which powertrain types tend to have more or fewer seats on average.



```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))
ax[0].bar(df['PowerTrain'],df['Segment'],color='red')
ax[1].bar(df['PowerTrain'],df['Seats'])

ax[0].set_title('Power vs segment')
ax[0].set_ylabel('Segment')


ax[1].set_title('Power vs seats')
ax[1].set_ylabel('Seats')
ax[1].set_xlabel('Power')

fig.show()
```

**Tree Map**

The treemap shows a hierarchical structure starting with 'Brand' as the top-level categories. Each 'Brand' contains 'Model' subcategories, and within each 'Model,' there are further subdivisions based on 'Range_Km.' **Divide the data into two new data set for giving answer of following questions**

```python
df_1 = df.loc[df['PriceEuro'] <= 50000]
df_2 = df.loc[df['PriceEuro'] > 50000]
```

**Which vehicle has max range (km) under 50,000 Euros?**

```python
top_range_1 = df_1.sort_values(by= 'Range_Km', ascending= False)
top_range_1[['FullName', 'Range_Km', 'Battery_Pack Kwh' ,'PriceEuro', 'RapidCharge']]
```

**Which vehicle has max range(Km) costing more than 50,000 Euros?**

```python
range_2 = df_2.sort_values(by= 'Range_Km', ascending= False)
range_2[['FullName', 'Range_Km', 'Battery_Pack Kwh' ,'PriceEuro', 'RapidCharge']]
```

**Vehicles with best acceleration under 50,000 Euros**

```python
acceleration = df_1.sort_values(by= 'AccelSec')
acceleration[['FullName','AccelSec', 'Range_Km', 'Battery_Pack Kwh' ,'PriceEuro']]
```

**Vehicles with best acceleration costing over 50,000 Euro**

```python
acceleration_2 = df_2.sort_values(by= 'AccelSec')
acceleration_2[['FullName','AccelSec', 'Range_Km', 'Battery_Pack Kwh' ,'PriceEuro']]
```

**Number of vehicles produced by each brand**

```python
companies = df.groupby('Brand').count()
companies['Model'].sort_values(ascending = False)
```
